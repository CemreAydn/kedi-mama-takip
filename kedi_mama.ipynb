```python

def sayi_al(soru):
    while True:
        try:
            sayi = float(input(soru))
            if sayi <= 0:
                print("geçerli bir sayı girdiğinizden emin olunuz")
            else:
                return sayi
        except ValueError:
            print("geçerli bir sayı gidiğinizden emin olunuz.")
isim = input("Kedinizin ismi nedir? ")
paket = sayi_al("Elinizde kaç gram mama var? ")
gunmiktarnet = input(isim + " bir günde kaç gram yemesi lazım, biliyor musunuz? (Evet / Hayır) ")

while gunmiktarnet != "Evet" and gunmiktarnet != "Hayır":
    print("Lütfen sadece Evet veya Hayır yazınız.")
    gunmiktarnet = input(isim + " bir günde kaç gram yemesi lazım, biliyor musunuz? (Evet / Hayır) ")

if gunmiktarnet == "Hayır":
    input("Hesaplamak için soruları yanıtlayın, devam etmek için Enter'a basınız. ")

    kilo = sayi_al(soru="kediniz kaç kilogram?")
    print("Kediniz hangi yaşam evresinde?")
    print("1 = yavru (0-4 ay)")
    print("2 = yavru (4-12 ay)")
    print("3 = kısır yetişkin")
    print("4 = kısır olmayan yetişkin")
    print("5 = yaşlı")
    secim = input("Numara gir: ")

    if secim == "1":
        carpan = 3.0
    elif secim == "2":
        carpan = 2.0
    elif secim == "3":
        carpan = 1.2
    elif secim == "4":
        carpan = 1.4
    elif secim == "5":
        carpan = 1.2
    else:
        carpan = 1.2

    print("Mama türü: 0 = paketteki değeri gireceğim, 1 = yavru, 2 = yetişkin normal, 3 = kısır, 4 = diyet, 5 = yaş mama")
    mama_secim = input("Numara gir: ")

    kcal_tablosu = {
        "1": 4.3,
        "2": 3.9,
        "3": 3.4,
        "4": 3.1,
        "5": 1.0
    }

    if mama_secim == "0":
        while True:
            try:
                kcal_kg = float(input("Elinizdeki mamanın kilogram başına kalorisi nedir? "))

                if kcal_kg <= 0:
                    print("Geçerli bir sayı girdiğinizden emin olunuz.")
                else:
                    break
            except ValueError:
                print("Geçerli bir sayı girdiğinizden emin olunuz.")

        kcal_g = kcal_kg / 1000
    else:
      kcal_g = kcal_tablosu[mama_secim]

    gram = carpan * 70 * kilo ** 0.75 / kcal_g
    print("Önerilen günlük", round(gram, 1), "gram")
    gunluk = gram
else:
   gunluk = sayi_al("Bir günde kaç gram yemesi lazım? ")

gun = int(paket / gunluk)
print("Mamanız", gun, "gün boyunca size yetecektir.")
```
